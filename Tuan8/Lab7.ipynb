```python
import nltk
from nltk.tokenize import word_tokenize

print("--- Bài 4: Loại bỏ stopword khỏi văn bản ---")
text = "This is a simple sentence to demonstrate removing stopwords from a text."
words = word_tokenize(text)
stop_words_eng = set(stopwords.words('english'))

filtered_text = [w for w in words if w.lower() not in stop_words_eng]
print("Gốc:", words)
print("Lọc:", filtered_text)

print("\n--- Bài 5: Bỏ qua các stopword khỏi danh sách ---")
print("Số lượng stopword ban đầu:", len(stop_words_eng))
words_to_remove = {'not', 'no'}
custom_stop_words = stop_words_eng - words_to_remove
print("Số lượng stopword sau khi bỏ qua 'not' và 'no':", len(custom_stop_words))
```

```python
from nltk.corpus import wordnet

print("--- Bài 6: Tìm định nghĩa và ví dụ ---")
syns = wordnet.synsets("computer")
print("Từ:", syns[0].name())
print("Định nghĩa:", syns[0].definition())
print("Ví dụ:", syns[0].examples())

print("\n--- Bài 7: Từ đồng nghĩa và trái nghĩa của từ 'good' ---")
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print("Đồng nghĩa:", set(synonyms))
print("Trái nghĩa:", set(antonyms))
```

```python
print("--- Bài 9: So sánh sự giống nhau của 2 danh từ ---")
noun1 = wordnet.synset('cat.n.01')
noun2 = wordnet.synset('dog.n.01')
print("Độ tương đồng giữa 'cat' và 'dog':", noun1.wup_similarity(noun2))

print("\n--- Bài 10: So sánh sự giống nhau của 2 động từ ---")
verb1 = wordnet.synset('run.v.01')
verb2 = wordnet.synset('sprint.v.01')
print("Độ tương đồng giữa 'run' và 'sprint':", verb1.wup_similarity(verb2))
```

```python
from nltk.corpus import names
import random

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print("--- Bài 11: Số lượng và 10 tên nam/nữ đầu tiên ---")
print(f"Nam: {len(male_names)} tên. 10 tên đầu: {male_names[:10]}")
print(f"Nữ: {len(female_names)} tên. 10 tên đầu: {female_names[:10]}")

print("\n--- Bài 12: 15 kết hợp ngẫu nhiên ---")
labeled_names = ([(name, 'male') for name in male_names] +
                 [(name, 'female') for name in female_names])
random.shuffle(labeled_names)
for name, label in labeled_names[:15]:
    print(f"{name}: {label}")

print("\n--- Bài 13: Trích xuất ký tự cuối ---")
featuresets = [(name[-1], gender) for (name, gender) in labeled_names]
print("10 đặc trưng đầu tiên (Ký tự cuối, Nhãn):")
print(featuresets[:10])
```
