```python
### YÊU CẦU 1: Xác định và phân loại dữ liệu định tính và định lượng.

**Dữ liệu định lượng (Quantitative data):** Là các dữ liệu có thể tính toán số học. Bao gồm các cột điểm số của các năm: T1..N1, T2..N2, T6..N6, DH1, DH2, DH3 và các biến tạo thêm như TBM1, TBM2, TBM3, US_TBM1, US_TBM2, US_TBM3.

**Dữ liệu định tính (Qualitative data):** Là các dữ liệu phân loại, thuộc tính. Bao gồm: GT (Giới tính), DT (Dân tộc), KV (Khu vực), KT (Khối thi), và các biến tạo thêm phân loại như XL1, XL2, XL3, KQXT.
```

```python
### YÊU CẦU 2: Định nghĩa các thang đo phù hợp cho từng biến số

**Thang đo định danh (Nominal Scale):** Dùng để phân loại, không có tính thứ tự. Áp dụng cho: GT (Nam/Nữ), DT (Kinh/Khác...), KV (1, 2, 2NT, 3), KT (A, A1, B, C, D1...), KQXT (1, 0).

**Thang đo thứ bậc (Ordinal Scale):** Có tính phân loại và theo thứ tự, nhưng khoảng cách không đều nhau. Áp dụng cho xếp loại: XL1, XL2, XL3 (Y, TB, K, G, XS).

**Thang đo tỉ lệ (Ratio Scale):** Có đầy đủ tính chất của các thang đo trên và có điểm 0 tuyệt đối (0 điểm mang ý nghĩa là không có điểm nào). Áp dụng cho tất cả các biến điểm số: T1..N1, T2..N2, T6..N6, DH1, DH2, DH3, TBM..., US_TBM....
```

```python
import pandas as pd
import numpy as np
```

```python
# YÊU CẦU 3: Tải dữ liệu và in ra màn hình
df = pd.read_csv('dulieuxettuyendaihoc.csv')

print("--- 10 dòng đầu tiên của dữ liệu ---")
print(df.head(10))

print("\n--- 10 dòng cuối cùng của dữ liệu ---")
print(df.tail(10))
```

```python
# YÊU CẦU 4: Thống kê và xử lý dữ liệu thiếu cho biến DT
print("\n--- Khảo sát biến Dân tộc (DT) ---")
print("Số lượng dữ liệu thiếu của DT:", df['DT'].isnull().sum())
print("Bảng tần số biến DT (bao gồm cả giá trị thiếu):")
print(df['DT'].value_counts(dropna=False))

# Thay thế dữ liệu thiếu của biến DT bằng 0
df['DT'] = df['DT'].fillna(0)
```

```python
# YÊU CẦU 5: Thống kê và xử lý dữ liệu thiếu cho biến T1 bằng Mean
print("\n--- Khảo sát biến Toán lớp 10 (T1) ---")
print("Số lượng dữ liệu thiếu của T1:", df['T1'].isnull().sum())
print("Bảng tần số biến T1:")
print(df['T1'].value_counts(dropna=False))

# Thay thế dữ liệu thiếu của T1 bằng giá trị trung bình (Mean)
df['T1'] = df['T1'].fillna(df['T1'].mean())
```

```python
# YÊU CẦU 6: Xử lý dữ liệu thiếu cho các biến điểm số còn lại bằng Mean
score_columns = [
    'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]

# Dùng vòng lặp để điền Mean cho tất cả các cột điểm còn lại nếu có dữ liệu thiếu
for col in score_columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# YÊU CẦU 7: Tạo các biến TBM1, TBM2, TBM3
# Công thức: (T*2 + L + H + S + V*2 + X + D + N)/10
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10
```

```python
# YÊU CẦU 8: Tạo các biến xếp loại XL1, XL2, XL3 dựa trên TBM
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif 5.0 <= tbm < 6.5:
        return 'TB'
    elif 6.5 <= tbm < 8.0:
        return 'K'
    elif 8.0 <= tbm < 9.0:
        return 'G'
    else:
        return 'XS'

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
```

```python
# YÊU CẦU 9: Min-Max Normalization chuyển đổi điểm từ thang 10 sang thang 4
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4
```

```python
# YÊU CẦU 10: Tạo biến KQXT (Kết quả xét tuyển: Đậu = 1, Rớt = 0)
def ket_qua_xet_tuyen(row):
    kt = row['KT']
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']

    # Tính điểm trung bình xét tuyển theo từng khối
    if kt in ['A', 'A1']:
        diem_xt = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        diem_xt = (dh1 + dh2*2 + dh3) / 4
    else:
        diem_xt = (dh1 + dh2 + dh3) / 3

    return 1 if diem_xt >= 5.0 else 0

# Sử dụng apply trên từng dòng (axis=1) để tính KQXT
df['KQXT'] = df.apply(ket_qua_xet_tuyen, axis=1)
```

```python
# YÊU CẦU 11: Lưu dữ liệu sau khi xử lý xuống file csv
output_file = 'processed_dulieuxettuyendaihoc.csv'
df.to_csv(output_file, index=False, encoding='utf-8')

print(f"\nĐã hoàn thành! Dữ liệu được lưu thành công vào file '{output_file}'")
```
